### PROMPT Constructor `prompt_extract_author()`

Extract the Author of listed on a piece of text.

```python
# IMPORT the Prompt Function and the System Message
from gptquery.tools.tool_text_extraction.extract_authornames.prompts.default import (prompt_extract_author,
                                                                                     EXTRACTION_SYSTEM_MESSAGE)


sample_first_page_text = """Common Market Law Review 33: 51-68, 1996.
O 1996 Kluwer Academic Publishers. Printed in the Netherlands

THE EUROPEAN PARLIAMENT AND ASSOCIATION
COUNCIL DECISIONS: THE EXAMPLE OF DECISION 1/95
OF THE ECITURKEY ASSOCIATION COUNCIL
NANETTE A.E.M. NEUWAHL*

1. Introduction
On 6 March 1995 the European Union and Turkey decided to strengthen
their relations through the completion of a Customs Union.' The agree-
ment reached involves the adoption of an Association Council Decision
(No. 1/95)2 establishing the Customs Union, an Association Council
Resolution providing for the development of co-operation, and a Declara-
tion by which the European Union announces the resumption of financial
co-operation with Turkey. These measures will anchor Turkey firmly in
Europe. They will contribute to modernizing its economic legislation and
they will also open a significant market to the Community. Yet uncer-
tainty persists as regards the role of the European Parliament (EP) and its
impact on such agreements. It is difficult to assess the role of EP, espe-
cially as long as the status of Decision 1/95 remains unclear.
"""

# PRINT System message and USER MESSAGE
print(EXTRACTION_SYSTEM_MESSAGE)
print("*"*100)
print(prompt_extract_author(first_page_text=sample_first_page_text))
```

```python
# IMPORT the Prompt Function and the System Message
from gptquery.tools.tool_text_extraction.extract_authornames.prompts.default import (prompt_extract_author)
# LOAD cmlr_authors_sample
cmlr_authors = pd.read_parquet(ROOT_DIR / "data" / "uot25rev" / "cmlr_authors_sample.parquet")
# TEST Prompt constructing function
cmlr_authors["validate_prompt"] = cmlr_authors.apply(lambda row: prompt_extract_author(row["first_page_text"]), axis=1)
```
